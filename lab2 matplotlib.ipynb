2. Napisz program który utworzy wykresy liniowe danych finansowych firmy Alphabet Inc. w
okresie od 3 października 2016 r. do 7 października 2016 r. (plik z danymi fdata.csv)

```python
import pandas as pd

df = pd.read_csv("data/fdata.csv")

df["Date"] = pd.to_datetime(df["Date"], format="%m-%d-%y")

start_date = "2016-10-03"
end_date = "2016-10-07"
df_filtered = df[(df["Date"] >= start_date) & (df["Date"] <= end_date)]

plt.figure(figsize=(10, 5))
plt.plot(df_filtered["Date"], df_filtered["Open"], linestyle="-", label="Open")
plt.plot(df_filtered["Date"], df_filtered["High"], linestyle="-", label="High")
plt.plot(df_filtered["Date"], df_filtered["Low"], linestyle="-", label="Low")
plt.plot(df_filtered["Date"], df_filtered["Close"], linestyle="-", label="Close")

plt.legend()
plt.grid()

plt.show()
```

3. Napisz program w Pythonie, aby wykreślić dwie lub więcej linii w różnych stylach.
Przykładowe rozwiązanie:

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, linestyle="-", color="b", linewidth=2, marker="o", label="sin(x)")
plt.plot(x, y2, linestyle="--", color="r", linewidth=2, marker="s", label="cos(x)")
plt.plot(x, y3, linestyle=":", color="g", linewidth=2, marker="^", label="sin(x) * cos(x)")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Wykres kilku funkcji w różnych stylach")
plt.legend()
plt.grid()

plt.show()
```

5. Napisz program, który utworzy wykres składający się z kilku linii w różnych stylach i kolorach.

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, np.cos(x), label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.plot(x, np.tan(x), label='tan(x)', color='green', linestyle=':', linewidth=2)

plt.show()
```

6. Napisz program, który w ramach jednego obszaru roboczego utworzy kilka wykresów.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y1, color="blue", label="sin(x)")
axs[0, 0].set_title("sin(x)")
axs[0, 0].legend()

axs[0, 1].plot(x, y2, color="red", label="cos(x)")
axs[0, 1].set_title("cos(x)")
axs[0, 1].legend()

axs[1, 0].plot(x, y3, color="green", label="tan(x)")
axs[1, 0].set_title("tan(x)")
axs[1, 0].set_ylim(-10, 10)
axs[1, 0].legend()

axs[1, 1].scatter(np.random.rand(50), np.random.rand(50), color="purple", label="scatter")
axs[1, 1].set_title("Random Scatter")
axs[1, 1].legend()

for ax in axs.flat:
    ax.grid()

plt.tight_layout()
plt.show()
```

13. Napisz program, który utworzy wykres słupkowy dla podanych danych. Na wykresie dane
mają być rozdzielone ze względu na płeć.

```python
categories = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width / 2, men_means, width, label='Men', color='green')
plt.bar(x + width / 2, women_means, width, label='Women', color='red')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores person')
plt.xticks(x, categories)
plt.legend()
plt.show()
```

15. Napisz program, który utworzy skumulowany wykres słupkowy dla podanych danych:
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4

data =   [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477],
[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]]

```python
people = ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8')
segments = 4
data = np.array([[3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477],
                 [11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
                 [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
                 [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]])

fig, ax = plt.subplots(figsize=(10, 7))
x = np.arange(len(people))
left = np.zeros(len(people))

row_sums = np.sum(data, axis=0)
for i in range(segments):
    ax.barh(x, data[i], left=left, label=f'Segment {i + 1}')
    for j, value in enumerate(data[i]):
        percentage = (value / row_sums[j]) * 100
        ax.text(left[j] + value / 2, j, f'{percentage:.1f}%', ha='center', va='center', fontsize=15, color='white')
    left += data[i]

ax.set_yticks(x)
ax.set_yticklabels(people)
ax.set_xlabel('Scores')

ax.legend()
plt.tight_layout()
plt.show()
```

16. Napisz program, który utworzy wykres słupkowy, w którym słupki wypełnione będą
teksturami:

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
patterns = ['/', '\\', '|', '-', '+', 'x']

bars = plt.bar(languages, popularity, color='skyblue', edgecolor='black')

for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages\nWorldwide, Oct 2017 compared to a year ago')
plt.show()
```

17. Napisz program tworzący wykres kołowy dla danych z popularnością języków
programowania.

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Znajdowanie indeksu największego kawałka
explode = [0.2 if value == max(popularity) else 0 for value in popularity]

plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140, explode=explode)
plt.axis('equal')
plt.show()
```
