4. Dodaj niestandardowe informacje do skryptu Bokeh.
Uzupełnij funkcję tak, aby dodawała niestandardowe informacje na wykresie
punktowym.

```python
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.io import output_notebook

output_notebook()
data = {
    "x": [1, 2, 3, 4, 5],
    "y": [6, 7, 2, 4, 5],
    "radius": [10, 12, 8, 15, 9]
}

source = ColumnDataSource(data)

p = figure(tools="hover", title="Scatter Plot with Hover", tooltips=[
    ("Index", "$index"),
    ("(x, y)", "($x, $y)"),
    ("Radius", "@radius")
])

p.scatter("x", "y", size="radius", source=source)

show(p)
```

6. Stwórz wykres słupkowy z możliwością interakcji.
Skonfiguruj wyświetlenie interaktywnego wykresu słupkowego za pomocą
ColumnDataSource.

```python
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

fruits = ['Apples', 'Oranges', 'Pears', 'Bananas']
counts = [10, 20, 14, 22]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))
p = figure(x_range=fruits, title="Interactive Bar Plot")
p.vbar(x='fruits', top='counts', width=0.6, source=source,
       fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

7. Zadanie z Wykresem Danych Czasowych (Time Series Plot):
o Dane: Wygeneruj losowe dane dotyczące wartości akcji w ciągu ostatnich 30 dni,
przyjmujące wartości w przedziale od 100 do 200.
o Stwórz wykres liniowy pokazujący zmiany wartości akcji w czasie. Dodaj do wykresu
linię trendu, a także interaktywne znaczniki, które pokazują datę oraz wartość akcji po
najechaniu kursorem na punkt.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_notebook
from bokeh.models.tools import CrosshairTool

output_notebook()

np.random.seed(42)
dates = pd.date_range(end=pd.Timestamp.today(), periods=30, freq='D')
prices = np.random.uniform(100, 200, size=30)

df = pd.DataFrame({'date': dates, 'price': prices})

df['trend'] = df['price'].rolling(window=7, min_periods=1).mean()

source = ColumnDataSource(df)

p = figure(x_axis_type="datetime", title="Zmiany wartości akcji w czasie",
           height=400, width=700, tools="pan, wheel_zoom, reset")

p.line('date', 'price', source=source, line_width=2, legend_label="Cena akcji", color="navy")

p.line('date', 'trend', source=source, line_width=2, legend_label="Linia trendu", color="red", line_dash="dashed")

p.scatter('date', 'price', size=6, source=source, color="navy", alpha=0.6)

hover = HoverTool(
    tooltips=[
        ("Data", "@date{%F}"),
        ("Cena", "@price{0.2f}")
    ],
    formatters={"@date": "datetime"},
    mode="vline"
)
p.add_tools(hover)

p.add_tools(CrosshairTool())

p.xaxis.axis_label = "Data"
p.yaxis.axis_label = "Cena akcji"
p.legend.location = "top_left"

show(p)
```

8. Zadanie z Wykresem Słupkowym (Bar Chart):
o Dane: Wygeneruj losowe dane dotyczące liczby sprzedanych produktów w różnych
kategoriach (np. Elektronika, Odzież, Żywność, Meble) dla losowych wartości od 50 do 500.
o Stwórz wykres słupkowy pokazujący liczbę sprzedanych produktów w każdej kategorii.
Dodaj etykiety do osi oraz tytuł wykresu. Wprowadź możliwość wyboru wybranej
kategorii, aby zobaczyć dokładne dane dotyczące sprzedaży.

```python
import numpy as np
import plotly.graph_objects as go

np.random.seed(42)

categories = ['Elektronika', 'Odzież', 'Żywność', 'Meble']
days = np.arange(1, 11)

sales_data = {category: np.random.randint(50, 500, size=len(days)) for category in categories}

color_map = {
    "Elektronika": "blue",
    "Odzież": "red",
    "Żywność": "green",
    "Meble": "purple"
}

fig = go.Figure()

for category in categories:
    fig.add_trace(go.Bar(
        x=days,
        y=sales_data[category],
        name=category,
        marker=dict(color=color_map[category]),
        visible=True
    ))

visibility_options = [[(i == j) for i in range(len(categories))] for j in range(len(categories))]

dropdown_options = [
    {"label": "Wszystkie", "method": "update",
     "args": [{"visible": [True] * len(categories)}, {"title": "Sprzedaż produktów"}]}
]

for i, category in enumerate(categories):
    dropdown_options.append({
        "label": category,
        "method": "update",
        "args": [
            {"visible": visibility_options[i]},
            {"title": f"Sprzedaż: {category}"}
        ]
    })

fig.update_layout(
    updatemenus=[{
        "buttons": dropdown_options,
        "direction": "down",
        "showactive": True
    }],
    title="Sprzedaż produktów w różnych dniach",
    xaxis_title="Dzień",
    yaxis_title="Sprzedaż"
)

fig.show()
```

9. Zadanie z Wykresem Różnicowym (Waterfall Chart):
o Dane: Wygeneruj losowe dane dotyczące miesięcznych przychodów i wydatków przez
rok (np. przychody i wydatki od stycznia do grudnia).
o Opracuj wykres kaskadowy ilustrujący przychody i wydatki oraz ich wpływ na zysk
netto. Umożliw zmianę miesięcznych wartości przychodów lub wydatków za pomocą
suwaków i obserwuj, jak zmienia się wynik finansowy.

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

months = ["Styczeń", "Luty", "Marzec", "Kwiecień", "Maj", "Czerwiec",
          "Lipiec", "Sierpień", "Wrzesień", "Październik", "Listopad", "Grudzień"]

revenues = np.random.randint(3000, 10000, size=len(months))
expenses = np.random.randint(2000, 8000, size=len(months))

revenue_sliders = [widgets.IntSlider(value=revenues[i], min=3000, max=10000, step=100,
                                     description=f"Przychód {months[i]}", continuous_update=True)
                   for i in range(len(months))]

expense_sliders = [widgets.IntSlider(value=expenses[i], min=2000, max=8000, step=100,
                                     description=f"Wydatki {months[i]}", continuous_update=True)
                   for i in range(len(months))]


def update_waterfall(**kwargs):
    revenue_values = [kwargs[f'revenue_{i}'] for i in range(len(months))]
    expense_values = [kwargs[f'expense_{i}'] for i in range(len(months))]

    net_profit = np.array(revenue_values) - np.array(expense_values)

    cumulative = np.cumsum([0] + list(net_profit[:-1]))

    plt.figure(figsize=(12, 6))
    colors = ['green' if p >= 0 else 'red' for p in net_profit]
    plt.bar(months, net_profit, bottom=cumulative, color=colors)

    plt.xlabel("Miesiące")
    plt.ylabel("Zysk netto")
    plt.title("Interaktywny wykres kaskadowy (Waterfall Chart)")
    plt.xticks(rotation=45)
    plt.axhline(0, color="black", linewidth=0.8)
    plt.show()


interactive_kwargs = {}
for i, slider in enumerate(revenue_sliders):
    interactive_kwargs[f'revenue_{i}'] = slider
for i, slider in enumerate(expense_sliders):
    interactive_kwargs[f'expense_{i}'] = slider

interactive_plot = widgets.interactive(update_waterfall, **interactive_kwargs)
display(interactive_plot)
```

10. Zadanie z Wykresem Kółkowym (Pie Chart):
o Dane: Wygeneruj losowe dane przedstawiające udział różnych kategorii wydatków
(np. Wynajem, Żywność, Transport, Rozrywka) w całkowitym budżecie z wartościami
od 10% do 40%.
o Zadanie: Stwórz wykres kołowy ilustrujący te udziały. Dodaj interaktywność, aby po
najechaniu na fragment koła pokazać dokładny procent oraz wartość w złotówkach.

```python
import numpy as np
import plotly.express as px
import plotly.io as pio
import plotly.offline as pyo

# Inicjalizacja trybu offline w Plotly
pyo.init_notebook_mode(connected=True)

pio.renderers.default = "notebook"

# Ustawienie seed dla powtarzalności wyników
np.random.seed(42)

# Kategorie wydatków
categories = ["Wynajem", "Żywność", "Transport", "Rozrywka"]

# Generowanie losowych procentowych udziałów (10%-40%) i normalizacja do 100%
percentages = np.random.randint(10, 41, size=len(categories))
percentages = (percentages / percentages.sum()) * 100

# Całkowity budżet (np. 5000 zł)
total_budget = 5000

# Obliczenie wartości w złotówkach
values = (percentages / 100) * total_budget

# Tworzenie wykresu w Plotly
fig = px.pie(
    names=categories,
    values=values,
    title="Udział kategorii wydatków w budżecie",
    hover_data={'Wartość (zł)': values, 'Procent': percentages},
    hole=0.3  # Efekt "donut chart"
)

# Formatowanie tooltipów
fig.update_traces(textinfo="percent+label", hovertemplate="%{label}: %{percent} (%{customdata[0]:.2f} zł)")

# Wyświetlenie wykresu
fig.show()
```
